# Distancias

```python
from scipy.spatial import distance_matrix
import pandas as pd
```

```python
data = pd.read_csv("../datasets/movies/movies.csv", sep = ";")
```

```python
data.head()
```

```python
movies = data.columns.tolist()[1:]
movies
```

```python
dd1 = distance_matrix(data[movies], data[movies], p = 1)
dd2 = distance_matrix(data[movies], data[movies], p = 2)
dd10 = distance_matrix(data[movies], data[movies], p = 10)
```

```python
def dm_to_df(dd,col_name):
    import pandas as pd
    return pd.DataFrame(dd,index = col_name,columns = col_name)
```

```python
#Matriz de distancias de manhatan
dm_to_df(dd1,data['user_id'])
```

```python
#Matriz de distancias de euclideana
dm_to_df(dd2,data['user_id'])
```

```python
#Matriz de distancias de minkowski p = 10
dm_to_df(dd10,data['user_id'])
```

# Clustering Jerárquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], 'ward')
Z
```

```python
plt.figure(figsize = (25,10))
plt.title('Dendogram jerarquico para el clustering')
plt.xlabel('ID de los usuarios de NETFLIX')
plt.ylabel('Distancia')
dendrogram(Z,leaf_rotation = 90, leaf_font_size = 15)
plt.show()
```

```python
Z = linkage(data[movies], 'average')
Z
```

```python
plt.figure(figsize = (25,10))
plt.title('Dendogram jerarquico para el clustering')
plt.xlabel('ID de los usuarios de NETFLIX')
plt.ylabel('Distancia')
dendrogram(Z,leaf_rotation = 90, leaf_font_size = 15)
plt.show()
```

```python
data[movies]
```

```python
Z = linkage(data[movies], 'complete')
Z
plt.figure(figsize = (25,10))
plt.title('Dendogram jerarquico para el clustering')
plt.xlabel('ID de los usuarios de NETFLIX')
plt.ylabel('Distancia')
dendrogram(Z,leaf_rotation = 90, leaf_font_size = 15)
plt.show()
```

```python
Z = linkage(data[movies], 'single')
Z
plt.figure(figsize = (25,10))
plt.title('Dendogram jerarquico para el clustering')
plt.xlabel('ID de los usuarios de NETFLIX')
plt.ylabel('Distancia')
dendrogram(Z,leaf_rotation = 90, leaf_font_size = 15)
plt.show()
```

```python
Z = linkage(data[movies], 'single',metric = 'cityblock')
Z
plt.figure(figsize = (25,10))
plt.title('Dendogram jerarquico para el clustering')
plt.xlabel('ID de los usuarios de NETFLIX')
plt.ylabel('Distancia')
dendrogram(Z,leaf_rotation = 90, leaf_font_size = 15)
plt.show()
```
