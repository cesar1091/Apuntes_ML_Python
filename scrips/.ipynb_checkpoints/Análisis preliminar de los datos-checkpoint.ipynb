## Missing Values

```python
pd.isnull(data['body']).values.ravel().sum()
```

Los valores que faltan en un data set puede venir por dos razones:
* Extraccion de los datos
* Recolección de los datos

### Borrado de valores que faltan

```python
# axis = 0 borra filas, axis = 1 borra columnas
# how = 'all' borra las filas que tienen todo NaN
# how = 'any' borra las filas si tienen al menos un NaN
data.dropna(axis = 0, how = 'all')               
```

```python
column_name=data.columns.values.tolist()
```

```python
column_name
```

```python
data.drop(['sex'],axis = 1,inplace = True)
```

```python
data = pd.concat([data,dummy_sex], axis = 1)
data.head()
```

```python
def createDummies(df, var_name):
    dummy = pd.get_dummies(data[var_name], prefix = var_name)
    df = df.drop([var_name], axis = 1)
    df = pd.concat([df,dummy], axis = 1)
    return df
```
